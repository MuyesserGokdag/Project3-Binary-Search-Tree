# **Binary Search Tree**
    [7, 5, 1, 8, 3, 6, 0, 9, 4, 2] dizisinin Binary-Search-Tree aşamaları yazılacak.

* **Aşamaları:**

      1) Dizi rastgele sıralanmış olduğundan ilk eleman root node olarak belirlenir. (sıralı dizi olsaydı ortadaki eleman seçilirdi)
          - Ve sırada olan elemanları peşi sıra roota göre yerleştirilir.
          - Küçük olan sayılar root'un soluna, büyük olan sayılar root'un sağına yazılır.
          - root: 7 (olarak belirlendi)

      2) İkinci sıradaki 5, root'tan küçük. Bu nedenle root'un soluna yazılır
          -->                7
                            /
                           5     
      3) 1 elemanı 7'den küçüktür ve 5'ten de küçüktür. 5'in soluna yazılır.
          -->                7
                            / 
                           5
                          /
                         1    
      4) 8 elemanı 7'den büyüktür, root'un sağına yazılır
          -->                7
                            / \
                           5   8
                          /
                         1  
      5) 3 elemanı 7'den küçük, root'un sol dalından devam edilmeli. 5'ten de küçük ama 1'den büyük, 5'in sağına yazılır.
          -->                7
                            / \
                           5   8
                          / 
                         1   
                        / \
                           3
      6) 6 elemanı 7'den küçük, 5'ten büyüktür. 5'in soluna yazılır.
          -->                7
                            / \
                           5   8
                          / \
                         1   6
                        / \
                           3
      7) 0 elemanı 7'den küçük, 5 ve 1'den de küçüktür. 1'in soluna yazılır. 
          -->                7
                            / \
                           5   8
                          / \
                         1   6
                        / \
                       0   3
      8) 9 elemanı 7'den büyük sağ dala geçtik, 8'den de büyüktür. 8'in sağına yazılır.
          -->                7
                            / \
                           5   8
                          / \   \
                         1   6   9
                        / \
                       0   3
      9) 4 elemanı 7'den küçük, 5'ten küçük 3'ten büyük. 3'ün sağına yazılır.
          -->                7
                            / \
                           5   8
                          / \   \
                         1   6   9
                        / \
                       0   3
                            \
                             4
      10) 2 elemanı 7'den küçük, 3'ten de küçük. 3'ün soluna yazılır ve dizi son halini alır.
          -->                7
                            / \
                           5   8
                          / \   \
                         1   6   9
                        / \
                       0   3
                          / \
                         2   4
